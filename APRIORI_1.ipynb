```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import csv
from apyori import apriori
```

```python
data=pd.read_csv('D:/4sem/DM/lab/store_data.csv',header=None)
data=data.fillna('')
data.head(100)
```

```python

record=[]
for i in range(0,7501):
        record.append([str(data.values[i,j]) for j in range(0,20) ])
        
record = [list(filter(None, lst)) for lst in record]
```

```python

association_rules=apriori(record,min_support=0.002,min_confidence=0.2,min_lift=3,min_length=3)
#association_rules = apriori(records, min_support=0.0045,min_confidence=0.2, min_lift=3, min_length=2)
asoc_rule=list(association_rules)
for i in range(0,20):
    print(asoc_rule[i])
```

```python
for item in asoc_rule:
# first index of the inner list
# Contains base item and add item
    
    pair = item[0]
    items = [x for x in pair]
    print("Rule:"+ items[0] + "-->"  + items[1])
#second index of the inner list
    print("Support:" + str(item[1]))
#third index of the list located at 0th
#of the third index of the inner list
    print("Confidence:" + str(item[2][0][2]))
    print("Lift:"  + str(item[2][0][3]))
    print("=====================================")
```

```python
strong=max(asoc_rule) 
print(strong)
```
